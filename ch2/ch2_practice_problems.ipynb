```python
# Question 9 - Auto Data
auto = pd.read_csv("../datasets/Auto.csv")
auto = auto.drop(auto[auto.values == '?'].index)

datatypes = {'quant': ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration'],
             'qual': ['origin', 'name']}

auto_quant = auto[datatypes['quant']].astype(np.float_)
```

```python
for (column_name, column_data) in auto_quant.items():
    print(f"{column_name}: range is {np.min(column_data):.2f} to {np.max(column_data):.2f} and mean is {np.mean(column_data):.2f} and stddev is {np.std(column_data):.2f}")
    
```

```python
auto_quant_reset = auto_quant.reset_index(drop=True)
auto_quant_dropped = auto_quant_reset.drop(range(10, 86))

for (column_name, column_data) in auto_quant_dropped.items():
    print(f"{column_name}: range is {np.min(column_data):.2f} to {np.max(column_data):.2f} and mean is {np.mean(column_data):.2f} and stddev is {np.std(column_data):.2f}")
    
```

```python
pd.plotting.scatter_matrix(auto, figsize=(20, 20));
```
